# 비원형 데이터의 군집화 - DBSCAN
---
- 분활적 군집화 방식
- 밀도(데이터의 밀집)기반 군집화 => 미리 군집수 지정 필요 없음
- 다양한 형태의 데이터에서 군집화 가능
- 이상치 데이터 제거도 가능함
- 군집을 정하는 기준
    - 임의의 점(Point)에서 지정된 거리만큼 영역안에 지정된 데이터 수 존재 여부
    - 하이퍼 파라미터 => 거리, 데이터수

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import pandas as pd
```

### [1] 데이터 로딩

```python
data, target = load_iris(return_X_y=True, as_frame=True)
```

```python
data.head(5)
```

### [2] 데이터 전처리

```python
scaler = StandardScaler()
scaler.fit(data)
scaler.transform(data)
```

### [3] 군집화

```python
# 군집화 객체 생성
dbscan = DBSCAN(eps=0.6, min_samples=8)
```

```python
# iris 데이터를 분류
irisDBS = dbscan.fit(data)
```

```python
print(irisDBS.labels_)
print(irisDBS.components_)
```

```python
pre = pd.Series(irisDBS.fit_predict(data))
compare = pd.concat((pre, target), axis=1)
compare.rename(columns={0:'dbscan_cluster'}, inplace=True)
```

```python
compare.groupby(['target'])['dbscan_cluster'].value_counts()
```
