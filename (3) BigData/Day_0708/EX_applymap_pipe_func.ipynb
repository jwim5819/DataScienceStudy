```python
# 모듈 로딩
import pandas as pd
import numpy as np

# 파일 관련 변수 선언
DIR_PATH = '../Data/'
FILE = DIR_PATH + 'titanic.csv'
```

```python
# DF객체 생성
titanicDF = pd.read_csv(FILE)
titanicDF.head()
```

```python
subDF = titanicDF.loc[:5,['Age','Fare']]
subDF['Value'] = 10
subDF
```

```python
# 사용자 정의 함수
def addValue(x):
    print(f'[Debug]-------------\n{x}')
    return x+x

def addTwoValue(x, y):
    print(f'[Debug]-------------\n{x}, {y}')
    return x+y

def missing_value_check(sr):
    print(f'[Debug]---------------\n{sr}')
    return sr.isnull()

def missing_value_sum(sr):
    print(f'[Debug]---------------\n{sr}')
    return sr.isnull().sum()
```
